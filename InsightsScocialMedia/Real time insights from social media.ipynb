##### I. The most important social media platform : Twitter
* Twitter has a huge influence in the world. It's a gold mine in term of information. It conveys the big waves of thoughts and moods sweep around the world and reflect almost on any subjects that happen in real time.
* Twitter provides both global and local trends. What we're doing here is try to find some most interesting trends in the data, which language is used frequently and in which countries.

```python
# Loading json module
import json
WW_trends = json.loads(open('datasets/WWTrends.json').read())
US_trends = json.loads(open('datasets/USTrends.json').read())
# Inspecting data by printing out ww_trends and us_trends
print(WW_trends)
print(US_trends)
```

##### II. Prettifying the output
* Tweets that downloaded from Twitter are hard to read. Using json.dumps() method to format thems as json string.

```python
# prettifying the results downloaded from twitter
print('WW_trends:'.format(json.dumps(WW_trends, indent=1)))
print('US_trends:'.format(json.dumps(US_trends, indent=1)))
```

##### III. Finding common trends
* Use python's set data structure to find common trends
* Iterate through the two trends objects, casts the list of names
* Use intersection method to find the common names between two sets

```python
# Extracting all the world wide trend names from the WW_trends
world_trends = set([trend['name'] for trend in WW_trends[0]['trends']])
# Extracting all the trend namses from the US_trends
us_trends = set([trend['name'] for trend in US_trends[0]['trends']])
# Intersect the two trends to find common trend names
common_trends = world_trends.intersection(us_trends)
# Inspecting the data
print('1. Size of world trends:',len(world_trends), '\n')
print(world_trends, '\n')
print('2. Size of us trends:',len(us_trends), '\n')
print(us_trends, '\n')
print('Size of Common Trends:',len(common_trends), '\n')
print('3. Common Trends:', common_trends)
```

##### IV. Exploring the hot trends
* Out of two sets of trends ( 50 in size for each trend), there're 11 overlapping topics with the most trending topic: WE LOVE THE EARTH
* With the popular hashtag #WeLoveTheEarth as query parameter, we've queried and got response from the search API stored in the dataset folder as 'WeLoveTheEarth.json'.

##### V. Digging Deeper
* Digging deeper into a tweet, let us realized that there're more to a tweet than just a short text as what we normally think. It is actually a long list with its elements as dictionaries that include : time, id, text, user name, screen name, location, geography,etc...
* The tweet was created on Friday April 19th, 2019 at 8:46 AM
* Let's focus on a few interesting fields and see if we can find hidden insights here

```python
# Extracting the text of all the tweets from the tweet object
texts = [tweet['text'] for tweet in tweets]
# Extracting all the names who tweet about this topic: We love the earth
names = [user_mentions['screen_name'] for tweet in tweets for user_mentions in tweet['entities']['user_mentions']]
# Extract all the hashtags being used when talking about this topic
hashtags = [hashtags['text'] for tweet in tweets for hashtags in tweet['entities']['hashtags']]
# Inspecting the first few results
print('Twitter texts regarding topic:- We Love the Earth', '\n')
print(json.dumps(texts[0:10], indent=1),'\n')
print('Names of persons texting this topic:- We Love the Earth', '\n')
print(json.dumps(names[0:10], indent=1),'\n')
print('Hashtags being used regarding topic:- We Love the Earth', '\n')
print(json.dumps(hashtags[0:10], indent=1),'\n')
```

##### VI. Frequency Analysis
* From the last results, we can deduce that:
    * We're talking about a song about loving the Earth
    * Many big name artists are the forces behind this twitter wave: Lil Dicky, Justin Bieber, Ariana Grande, Ed Sheeran, etc...
* Observing the first 10 fields gives us a sense of these particular data
* Let's computing data frequency distribution and decide what we'll do next

```python
from collections import Counter
# counting occurencies of text, getting a list of all names and hashtags associate with it
for item in [names, hashtags]:
    c = Counter(item)
    print(c.most_common(10), '\n')
```

##### VII. Activity around the trend
* By now we can positively deduce that this was about a music video about the Earth ('Earth Music Video' by Lil Dicky)
* Leo DiCaprio isn't an artist, but he's an environmentalist; so it's quite undestandable why his name was there
* Let's see how namy retweets around this trend. A popularity's tweet is based on how people react to that tweet - retweets



```python
# Extracting useful information from retweets
retweets = [(tweet['retweet_count'],
             tweet['retweeted_status']['favorite_count'],
             tweet['retweeted_status']['user']['followers_count'],
             tweet['retweeted_status']['user']['screen_name'],
             tweet['text']) for tweet in tweets if 'retweeted_status' in tweet]
```

##### VIII. Create a table that speaks a 1000 words
* From the data we obtained from this trend, we can build a data frame that is well structured and easier to read

```python
# Importing modules
import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame and visualize the data in a pretty and insightful format
df = pd.DataFrame(retweets,
                  columns=['Retweets','Favorites', 'Followers', 'ScreenName', 'Text']).groupby(
    ['ScreenName','Text','Followers']).sum().sort_values(by='Followers', ascending=False)
df.style.background_gradient()
```

##### IX. Analyzing Used Languages
* Lil Dicky has the most responses from his followers — 42.4% of his followers liked his first tweet.
* Even if celebrities like Katy Perry and Ellen have a huuge Twitter following, their followers hardly reacted, e.g., only 0.0098% of Katy's followers liked her tweet.
* While Leo got the most likes and retweets in terms of counts, his first tweet was only liked by 2.19% of his followers.
* The large differences in reactions could be explained by the fact that this was Lil Dicky's music video. Leo still got more traction than Katy or Ellen because he played some major role in this initiative.

```python
# Extracting language for each tweet and appending it to the list of languages
tweets_languages = []
for tweet in tweets: 
    tweets_languages.append(tweet['lang'])
    
# Plotting the distribution of languages
import matplotlib.pyplot as plt
%matplotlib inline
plt.grid()
plt.hist(tweets_languages)
plt.title('Languages Used')
plt.xlabel('Languages')
plt.ylabel('Counts')
plt.show()
```
